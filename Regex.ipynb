```python
import csv
import re
import random
import os
```

This block imports the libraries we need:
- csv: for reading and writing CSV files
- re: for using regular expressions
- random: for generating random names, courses, and sections
- os: for checking if files already exist

```python
def generate_student_data(num_records=100000):
    first_names = ['Aarav', 'Vivaan', 'Aditya', 'Vihaan', 'Arjun', 'Sai', 'Reyansh', 'Ayaan', 'Krishna', 'Ishaan', 'Saanvi', 'Aanya', 'Aadhya', 'Kiara', 'Diya', 'Pari', 'Ananya', 'Riya', 'Siya', 'Fatima']
    last_names = ['Sharma', 'Verma', 'Gupta', 'Singh', 'Kumar', 'Patel', 'Reddy', 'Agarwal', 'Jain', 'Khan', 'Mehta', 'Shah', 'Iyer', 'Menon', 'Nair']

    course_codes = [
        'CS3126', 'CS3202',
        'EE3101', 'EE3205'
    ]

    sections = ['AI1', 'AI2', 'AI3', 'CS1', 'CS2', 'CS3']

    filename = 'student_records.csv'

    if os.path.exists(filename):
        print(f"'{filename}' already exists. Skipping data generation.")
        return filename

    print(f"Generating {num_records} student records...")

    with open(filename, 'w', newline='') as csvfile:
        fieldnames = ['Student Name', 'Roll Number', 'Courses Taken', 'Email', 'Section']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        writer.writeheader()

        for i in range(1, num_records + 1):
            first = random.choice(first_names)
            last = random.choice(last_names)
            student_name = f"{first} {last}"

            roll_number = f"SE23UARI{i:03d}"

            num_courses = random.randint(3, 4)
            courses_taken = ','.join(random.sample(course_codes, num_courses))

            email = f"{first.lower()}.{last.lower()}{i}@mahindrauniversity.edu.in"

            section = random.choice(sections)

            writer.writerow({
                'Student Name': student_name,
                'Roll Number': roll_number,
                'Courses Taken': courses_taken,
                'Email': email,
                'Section': section
            })

    print(f"Successfully generated and saved data to '{filename}'")
    return filename
```

This block defines a function to generate random student records.
It creates student names, roll numbers, course lists, email addresses, and sections.
The data is saved into a file called student_records.csv.
If the file already exists, it skips regeneration.

```python
def perform_regex_matching(data_file):
    if not os.path.exists(data_file):
        print(f"Error: Data file '{data_file}' not found.")
        return

    with open(data_file, 'r') as csvfile:
        reader = csv.DictReader(csvfile)
        records = list(reader)

    print("\n--- Starting Regular Expression Matching ---\n")

    print("1. Finding all students in Section 'AI1'...")
    section_pattern = re.compile(r'^AI1$')
    ai1_students = [rec for rec in records if section_pattern.match(rec['Section'])]
    print(f"   Found {len(ai1_students)} students in Section AI1.")
    print("   --- Sample Matches ---")
    for student in ai1_students[:5]:
        print(f"   - Name: {student['Student Name']}, Roll No: {student['Roll Number']}")
    print("-" * 40)

    print("2. Finding all students who have taken course 'CS3126'...")
    course_pattern = re.compile(r'CS3126')
    cs_students = [rec for rec in records if course_pattern.search(rec['Courses Taken'])]
    print(f"   Found {len(cs_students)} students who took CS3126.")
    print("   --- Sample Matches ---")
    for student in cs_students[:5]:
        print(f"   - Name: {student['Student Name']}, Courses: {student['Courses Taken']}")
    print("-" * 40)

    print("3. Finding roll numbers ending between 100 and 199...")
    roll_no_pattern = re.compile(r'SE23UARI1\d{2}$')
    specific_roll_nos = [rec for rec in records if roll_no_pattern.match(rec['Roll Number'])]
    print(f"   Found {len(specific_roll_nos)} roll numbers in the specified range.")
    print("   --- Sample Matches ---")
    for student in specific_roll_nos[:5]:
        print(f"   - Name: {student['Student Name']}, Roll No: {student['Roll Number']}")
    print("-" * 40)

    print("4. Validating all email addresses...")
    email_pattern = re.compile(r'^[a-z]+\.[a-z]+\d+@mahindrauniversity\.edu\.in$')
    invalid_emails = [rec for rec in records if not email_pattern.match(rec['Email'])]

    if not invalid_emails:
        print("   All email addresses are in a valid format.")
    else:
        print(f"   Found {len(invalid_emails)} invalid email addresses.")
        print("   --- Sample Invalid Emails ---")
        for student in invalid_emails[:5]:
            print(f"   - Invalid Email: {student['Email']}")
    print("-" * 40)
```

This block defines a function that uses regular expressions to check the student data.
1. Finds students in section AI1.
2. Finds students who have taken course CS3126.
3. Finds roll numbers between 100 and 199.
4. Validates that all emails are correct.

```python
if __name__ == "__main__":
    student_data_file = generate_student_data(num_records=100000)
    perform_regex_matching(student_data_file)
```

This is the main block that runs the program. It first generates the data file
and then runs the regex matching function on it.
