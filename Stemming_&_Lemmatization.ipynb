
This notebook demonstrates the use of different stemming algorithms (Porter, Lancaster, Snowball) and lemmatization using NLTK.

## Importing required libraries

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize
from nltk.corpus import wordnet
```

## Tokenization
We start with a sample sentence and tokenize it into individual words.

```python
text = "The striped bats are hanging on their feet for best viewing from the running children."
tokens = word_tokenize(text)

print(f"Original Tokens: {tokens}\n")
print("-" * 50)
```

## Porter Stemmer

```python
porter = PorterStemmer()
porter_stemmed = [porter.stem(token) for token in tokens]
print("Porter Stemmer Output:")
print(porter_stemmed)
print("\n" + "-" * 50)
```

## Lancaster Stemmer (more aggressive)

```python
lancaster = LancasterStemmer()
lancaster_stemmed = [lancaster.stem(token) for token in tokens]
print("Lancaster Stemmer Output (more aggressive):")
print(lancaster_stemmed)
print("\n" + "-" * 50)
```

## Snowball Stemmer (English)

```python
snowball = SnowballStemmer(language='english')
snowball_stemmed = [snowball.stem(token) for token in tokens]
print("Snowball Stemmer Output (English):")
print(snowball_stemmed)
print("\n" + "=" * 50)
```

## WordNet Lemmatizer (without POS tags)
By default, the WordNet lemmatizer assumes nouns.

```python
lemmatizer = WordNetLemmatizer()

lemma_no_pos = [lemmatizer.lemmatize(token) for token in tokens]
print("Lemmatizer Output (without POS tags - assumes Noun):")
print(lemma_no_pos)
print("\n" + "-" * 50)
```

## WordNet Lemmatizer (with POS tags)
We use NLTK's POS tagging to improve accuracy of lemmatization.

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

pos_tags = nltk.pos_tag(tokens)
lemma_with_pos = [lemmatizer.lemmatize(token, get_wordnet_pos(pos)) for token, pos in pos_tags]
print("Lemmatizer Output (with POS tags - more accurate):")
print(lemma_with_pos)
print("\n" + "=" * 50)
```
